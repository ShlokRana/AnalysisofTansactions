## 2) On the basis of the Quantity.

```python
cust_df['QuantitySum'].describe()
```

Making the categories after analyzing the QuantitySum Column

```python
QuantityBin = []
for i in cust_df['QuantitySum']:
  if(1<=i<200):
    QuantityBin.append("Cat-1")
  if(200<=i<400):
    QuantityBin.append("Cat-2")
  if(400<=i<600):
    QuantityBin.append("Cat-3")
  if(600<=i<800):
    QuantityBin.append("Cat-4")
  if(800<=i<1000):
    QuantityBin.append("Cat-5")
  if(1000<=i<10000):
    QuantityBin.append("Cat-6")
  if(10000<=i<400000):
    QuantityBin.append("Cat-7")

print(QuantityBin)
```

```python
count=0
for i in QuantityBin:
  if(i == 'Cat-7'):
    count = count+1
print(count)
```

```python
cust_df['QuantityBin'] = QuantityBin
cust_df
```

```python
fig, ax = plt.subplots(figsize =(10, 8))

ax.bar(cust_df['QuantityBin'].value_counts().index, cust_df['QuantityBin'].value_counts().values, color='orange', linewidth=1)
plt.xticks(rotation='horizontal')
ax.set_title('Segmentation of customer on the basis of Quantity', loc ='left', )
ax.set_ylabel('Number of customers')
ax.set_xlabel('Quantity Categories')
   
plt.show()
```

# Analyzing loyal Customers

```python
loyal_df = df.groupby(['Customer ID']).agg({'Quantity': ['min', 'max', 'sum']}).reset_index()
loyal_df.columns = ['Customer ID', 'QuantityMin', 'QuantityMax', 'QuantitySum']
loyal_df.head(20)
```

```python
thresh = dict(df['Customer ID'].value_counts())
print(thresh)
```

Chainging the key and value of the dictionary to add in a dataframe

```python
count = 0
auxlist = []
for i in loyal_df['QuantitySum']:
  x = i/thresh[loyal_df['Customer ID'].loc[count]]
  auxlist.append(x)
  count = count+1

print(auxlist)
print(len(auxlist))
```

```python
loyal_df['thresh'] = auxlist
loyal_df
```

```python
loyal_df['thresh'].describe()
```

## We are making the hypothesis and generating the formula for selecting the Loyal Customers 
1. thresh formula = QuantitySum of a particular customer/Total Ocurrance of the Customer entry in a data
2. Adding the thresh column in a data frame
3. After analyzing the max, min, mean, and std and analyzing the customer count on different thresh values, we are making the hypothesis that more than 250 values of thresh would be our **Loyal Customer**

So, there are 57 **Loyal Customers**




```python
loyal_df[loyal_df['thresh']>250].shape
```

## Plotting the top 10 Loyal Customers

```python
loyal_plot = loyal_df[loyal_df['thresh']>250]
loyal_plot = loyal_plot.sort_values(by='thresh',ascending=False)
#converting the Customer ID column in str to plot a barplot
loyal_plot['Customer ID'] = loyal_plot['Customer ID'].astype(str)
loyal_plot = loyal_plot.head(10)
print(loyal_plot)
```

```python
fig, ax = plt.subplots(figsize =(10, 8))

ax.bar(loyal_plot['Customer ID'].value_counts().index, loyal_plot['thresh'], color='orange', linewidth=1)
plt.xticks(rotation='horizontal')
ax.set_title('Top 10 Loyal Customers', loc ='left', )
ax.set_ylabel('Value of a thresh')
ax.set_xlabel('Customer ID')
   
plt.show()
```

# Customers Lifetime

```python
life_df = year_df
life_df.drop(['Invoice', 'StockCode', 'Description', 'Quantity', 'Price', 'Country'], axis=1)
life_df = life_df.groupby(['Customer ID']).agg({'InvoiceDate': ['min', 'max']}).reset_index()
life_df.columns = ['Customer ID', 'MinDate', 'MaxDate']
```

```python
life_df['Life'] = (life_df['MaxDate'] - life_df['MinDate']).dt.days
life_df
```

```python
#Sorting the Dataframe on the basis of Life
life_df = life_df.sort_values(by='Life',ascending=False).reset_index()
life_df = life_df.drop(['index'], axis=1)
life_df
```

## Plooting the Top Ten Customers Life

```python
life_df = life_df.head(10)
life_df
```

```python
# For Plotting 
list1 = list(life_df['Customer ID'])
list2 = list(life_df['Life'])
for item in range(10):
  list1[item] = str(list1[item])

plt.figure(figsize = (10,6))
plots = sns.barplot(x=list1, y=list2)
plt.title("Customers Lifetime")
plt.xlabel('Customer ID')
plt.ylabel('Lifetime in Days')
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 5),
                   textcoords='offset points')

plt.show()
```
